```python
import gym
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PATH = "/Volumes/BC_Clutch/Dropbox/DeepRLND/rl_multiagent/"
```

```python
avg_reward = pd.read_csv(PATH + "results/global_avg_reward.csv")
reward = pd.read_csv(PATH + "results/global_reward.csv")
```

```python
def plot_reward(data, tag):
    plt.plot(data.Step, data.Value)
    plt.title(f"MADDPG Tennis: {tag}")
    plt.ylabel("Average Reward")
    plt.xlabel("Episode")
    plt.savefig(PATH + f"charts/{tag}.png")
```

### Tennis, MADDPG

Note that average reward is over the most recent 100 episodes. 

```python
plot_reward(avg_reward,"Average_Reward")
```

```python
plot_reward(reward,"Reward")
```
